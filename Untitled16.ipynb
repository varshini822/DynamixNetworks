```python
# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
# 2. Upload Dataset (netflix_titles.csv from Kaggle)
uploaded = files.upload()   # choose netflix_titles.csv
netflix = pd.read_csv("netflix_titles.csv")
```

```python
# 3. Basic Info
print("Dataset Shape:", netflix.shape)
print("\nColumns:", netflix.columns)
print("\nMissing Values:\n", netflix.isnull().sum())
print("\nSample Data:\n", netflix.head())
```

```python
# 4. Data Cleaning
netflix.drop_duplicates(inplace=True)
netflix['date_added'] = pd.to_datetime(netflix['date_added'], errors='coerce')
netflix['year_added'] = netflix['date_added'].dt.year
netflix['month_added'] = netflix['date_added'].dt.month

# Fill missing country with 'Unknown'
netflix['country'] = netflix['country'].fillna("Unknown")
```

```python
# 5. Exploratory Data Analysis & Visualizations

# Movies vs TV Shows
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=netflix, palette="Set2")
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
# Top 10 Countries by Content
plt.figure(figsize=(8,6))
netflix['country'].value_counts().head(10).plot(kind='bar', color="skyblue")
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# Content added over time
plt.figure(figsize=(10,6))
netflix['year_added'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title("Content Added Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```

```python
# Top 10 Directors
plt.figure(figsize=(10,6))
netflix['director'].dropna().value_counts().head(10).plot(kind='bar', color="orange")
plt.title("Top 10 Directors on Netflix")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# Top 10 Genres (listed_in column)
plt.figure(figsize=(12,6))
all_genres = netflix['listed_in'].str.split(',').explode().str.strip()
all_genres.value_counts().head(10).plot(kind='bar', color="green")
plt.title("Top 10 Genres on Netflix")
plt.ylabel("Count")
plt.show()
```

```python
# Rating distribution
plt.figure(figsize=(8,5))
netflix['rating'].value_counts().plot(kind='bar', color="purple")
plt.title("Distribution of Ratings")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# Movies duration distribution
movies = netflix[netflix['type']=="Movie"]
movies['duration_num'] = movies['duration'].str.replace(" min","").astype(float)

plt.figure(figsize=(10,5))
sns.histplot(movies['duration_num'], bins=30, kde=True, color="red")
plt.title("Distribution of Movie Durations")
plt.xlabel("Duration (minutes)")
plt.show()
```

```python
# TV Shows seasons distribution
tv = netflix[netflix['type']=="TV Show"]
tv['seasons'] = tv['duration'].str.replace(" Season","").str.replace("s","").astype(int)

plt.figure(figsize=(8,5))
sns.countplot(x='seasons', data=tv, palette="coolwarm", order=tv['seasons'].value_counts().index)
plt.title("Distribution of TV Show Seasons")
plt.xlabel("Number of Seasons")
plt.ylabel("Number of TV Shows")
plt.show()
```
